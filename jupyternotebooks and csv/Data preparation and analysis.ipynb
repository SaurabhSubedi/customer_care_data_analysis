# Calls

```python
calls = pd.read_csv('calls.csv',parse_dates = ['started_at','ended_at','created_at','updated_at','called_date'])
calls.info()
```

```python
temp_Call = calls[['served_time','turn_around_time']]
temp_Call
```

```python
calls.head()
```

```python
calls['turn_around_time'] = pd.to_timedelta(calls['turn_around_time'])
calls['served_time'] = pd.to_timedelta(calls['served_time'])
calls['waiting_time'] = pd.to_timedelta(calls['waiting_time'])
```

```python
def time_to_seconds(cell):
    return round(cell.total_seconds()/60,2)

calls['waiting_time'] = calls['waiting_time'].apply(time_to_seconds)
```

```python
calls['served_time'] = calls['served_time'].apply(time_to_seconds)
calls['turn_around_time'] = calls['turn_around_time'].apply(time_to_seconds)
```

```python
filling_values = {'served_time': calls['served_time'].mean(), 'turn_around_time': calls['turn_around_time'].mean()}
calls = calls.fillna(value = filling_values)
calls
```

```python
calls.info()
```

# Correlation between attributes

<span style="color:red">Using Karl Pearsons correlation coffecient</span>

```python
calls.head(20)
calls.info()
```

```python
subset1 = calls[['waiting_time','served_time']]
wait_serve_corr = subset1['waiting_time'].corr(subset1['served_time'])
wait_serve_corr
```

<span style="color:blue ;font-size:25px">Conclusion <hr></span>
Weak Relationship : the value is close to 0 so there is almost no relationship between waiting times and served time

Positive Vlaue:The positive coefficient indicates that  a slight increase in served time is observed with an increasing waiting time.
    
Finally these values Seem to be largely independent of each other
    

```python
wait_turnaround_corr = calls['waiting_time'].corr(calls['turn_around_time'])
wait_turnaround_corr
```

<span style="color:blue ;font-size:25px">Conclusion <hr></span>
Strong Relationship: The closer the value is to 1, there is strong relationship between waiting and turn_around time

Positive value:  This tells us that as waiting time goes up, turnaround times also goes up.

Finally,  this means  higher waiting times implies longer served time as well

# Peak Hours

```python
peak_hrs  = calls[['started_at','id']]
```

```python
peak_hrs['hours'] = peak_hrs['started_at'].dt.hour
peak_hrs
```

```python
peak_hrs = peak_hrs.drop('started_at',axis=1)
```

```python
grouped_by_hrs = peak_hrs.groupby('hours')
grouped_by_hrs.get_group(18)
```

```python
peak_hrs_final = grouped_by_hrs.size()
peak_hrs_final = peak_hrs_final.reset_index()
peak_hrs_final.columns =['time(in_Hrs)','no_of_calls']
peak_hrs_final.sort_values(by=['no_of_calls'],ascending=False)
```

```python
peak_hrs_final.plot(x='time(in_Hrs)',y='no_of_calls',kind='bar',color='purple',title="PEAK HOURS")
```

<span style = "color: green; font-size : 30px; ">It can be concluded that 1pm to 4 pm are the peak hours for calls </span>

# Services with higher wait time

```python
service_wait_time = calls[['service_id','waiting_time']]
service_wait_time
```

```python
def letter_converter(value):
    if value == 1:
        return 'WIS'
    elif value ==2:
        return 'New Registration'
    else:
        return 'Training'
    
service_wait_time['service_name'] = service_wait_time['service_id'].apply(letter_converter)
service_wait_time
```

```python
service_wait_time = service_wait_time.drop('service_id',axis=1)
service_wait_time
```

```python
grp_service_wait_time = service_wait_time.groupby('service_name')
service_wait_time_summarized = grp_service_wait_time.mean()
```

```python
service_wait_time_summarized= service_wait_time_summarized.reset_index()
service_wait_time_summarized.columns = ['service_name','average_wait_time(in_minutes)']
service_wait_time_summarized.sort_values(by = ['average_wait_time(in_minutes)'])
```

```python
service_wait_time_summarized.plot(x='service_name',y='average_wait_time(in_minutes)',kind='bar',color='purple',title="Service by Waiting time")
```

<span style = "color: green; font-size : 30px; ">Training Services has highest waiting time. </span>
